# **Creating the data.yaml or the config file for the yolov8 model**

```python
import yaml

data_yaml = dict(
    train ='/kaggle/working/naruto-hand-sign-3/train',
    val ='/kaggle/working/naruto-hand-sign-3/valid',
    test='/kaggle/working/naruto-hand-sign-3/test',
    names =  ['bird', 'boar', 'dog', 'dragon', 'hare', 'horse', 'monkey', 'ox', 'ram', 'rat', 'snake', 'tiger'],
    nc = 12,
)

with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
    
%cat data.yaml
```

# **Defining the YOLOv8 model and training it for 30 epochs**

```python
from ultralytics import YOLO
model = YOLO('yolov8m.pt')
model.train(data='data.yaml', epochs=30, imgsz=640, batch=32,exist_ok=True)
```
